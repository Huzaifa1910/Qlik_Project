```python
# codeList_rdd = sc.parallelize(codeList,3)
# print(codeList_rdd.glom().collect()[0])

def scrap_the_web(codeList):
    try:
        # return codeList
        page_sourceList = []
        # code_list=['US','PK']
        # chrome_options = Options()
        # chrome_options.add_argument('--headless')
        # chrome_options.add_argument('--no-sandbox')
        # chrome_options.add_argument('--disable-dev-shm-usage')
        # driver = webdriver.Chrome('/usr/bin/chromedriver',chrome_options=chrome_options)
        
        driver = webdriver.Chrome('C:/chromedriver.exe')
        # driver = webdriver.Chrome()
        driver.maximize_window()
        url = 'https://www.qlik.com/us/partners/find-a-partner'
        driver.get(url)
        time.sleep(5)
        select = Select(driver.find_element(By.ID, "zl_countryCode"))
        page_source = driver.page_source
        soup = BeautifulSoup(page_source, 'html.parser')

        select = soup.find('select', id='zl_countryCode')

        # Find all the option tags within the select element
        options = select.find_all('option')

        # Loop through the option tags and print their value attributes
        # for option in options:
        #     codeList.append(option.get('value'))
        #     print(option.get('value'))
        # print(codeList)
        for i in range(1,len(codeList)):
        # for i in range(1,10):
            select = Select(driver.find_element(By.ID, "zl_countryCode"))
            select.select_by_value(codeList[i])
            time.sleep(5)
            try:
                button = driver.find_element(By.ID, 'zl_show-more-btn')
            except:
                print("NO Button present")
            while True:
                time.sleep(2)
                try:
                    button.click()
                    WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.ID, "zl_show-more-btn")))
                except:
                    print("Can not click")
                    break
                div = driver.find_element(By.ID, "zl_show-more")
                if div.value_of_css_property("display") == "none":
                    print("done Scrapping")
                    break
            page_source = driver.page_source
            page_sourceList.append(page_source)
        # ti.xcom_push(key='page_source', value=page_sourceList)
        driver.quit()
        return page_sourceList
    except NoSuchElementException:
        print(f"Cannot locate option with value: {codeList}")
        return None
# scrap_the_web(codeList_rdd)
```

```python
def sql_dumping():
    # create a connection to the database
    mydb = mysql.connector.connect(
        host="localhost",
        user="root",
        password="",
        database="qlikprojspark"
    )

    for w in range(len(data)):
        print("chala")
        print(len(data))
        # create a cursor object
        mycursor = mydb.cursor()

        # define the data to be inserted
        dataList = data

        # define the SQL query to create a table
        # Provide the table name here
        table_name2 = ''
        print(dataList[w][0][-1])
        table_name = dataList[w][0][-1].split(' ')
        print(table_name)
        if(len(table_name) > 1):
            for i in range(len(table_name)):
                table_name2 += table_name[i]
                table_name2 +="_"
            print(table_name2)
        create_table_sql = f"CREATE TABLE {table_name2} (name VARCHAR(255), category VARCHAR(255), address VARCHAR(255), tier VARCHAR(255), country VARCHAR(255))"

        # execute the query to create the table
        mycursor.execute(create_table_sql)

        # define the SQL query to insert data into the table
        insert_data_sql = f"INSERT INTO {table_name2} (name, category, address, tier, country) VALUES (%s, %s, %s, %s, %s)"

        # execute the query to insert data into the table
        mycursor.executemany(insert_data_sql, dataList[w])

        # commit the changes to the database
        mydb.commit()

        # print the number of rows that were inserted
        print(mycursor.rowcount, "rows were inserted into the", table_name, "table.")

sql_dumping()
```
